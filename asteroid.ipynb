```python
import requests
import json
import time
import os

# === CONFIG ===
CACHE_FILE = "nasa_cache.json"
BATCH_FOLDER = "nasa_batches"
BATCH_SIZE = 100  # number of rows per batch
os.makedirs(BATCH_FOLDER, exist_ok=True)


# === LOAD CACHE ===
try:
    with open(CACHE_FILE, "r") as f:
        cache = json.load(f)
except FileNotFoundError:
    cache = {}


# === NASA API CALL ===
def fetch_asteroid_data(asteroid_id):
    """Fetch asteroid data from NASA SBDB API with caching and retries."""
    if asteroid_id in cache:
        return cache[asteroid_id]

    url = f"https://ssd-api.jpl.nasa.gov/sbdb.api?sstr={asteroid_id}"

    for attempt in range(3):
        try:
            response = requests.get(url, timeout=15)
            if response.status_code == 200:
                data = response.json()
                cache[asteroid_id] = data

                # save cache after each successful fetch
                with open(CACHE_FILE, "w") as f:
                    json.dump(cache, f)

                time.sleep(0.3)  # prevent rate-limit
                return data
        except requests.exceptions.RequestException:
            time.sleep(1)
    print(f"⚠️ Skipping {asteroid_id} after 3 failed attempts.")
    return None


# === FILL MISSING FIELDS ===
def fill_from_nasa(df):
    """Fill missing orbital fields using NASA SBDB API data."""
    for i, row in df.iterrows():
        asteroid_id = str(row['Name']).strip()
        data = fetch_asteroid_data(asteroid_id)
        if not data:
            continue

        try:
            elements = {elem["name"]: float(elem["value"]) for elem in data["orbit"]["elements"]}

            if pd.isna(row.get('Semi Major Axis')) and 'a' in elements:
                df.at[i, 'Semi Major Axis'] = elements['a']
            if pd.isna(row.get('Perihelion Arg')) and 'w' in elements:
                df.at[i, 'Perihelion Arg'] = elements['w']
            if pd.isna(row.get('Asc Node Longitude')) and 'om' in elements:
                df.at[i, 'Asc Node Longitude'] = elements['om']
            if pd.isna(row.get('Mean Anomaly')) and 'ma' in elements:
                df.at[i, 'Mean Anomaly'] = elements['ma']
            if pd.isna(row.get('Aphelion Dist')) and 'q' in elements:
                df.at[i, 'Aphelion Dist'] = elements['q']
            if pd.isna(row.get('Orbital Period')) and 'per' in elements:
                df.at[i, 'Orbital Period'] = elements['per']
            if 'i' in elements and ('Inclination' not in df.columns or pd.isna(row.get('Inclination'))):
                df.at[i, 'Inclination'] = elements['i']

        except Exception as e:
            print(f"Error parsing {asteroid_id}: {e}")
    return df


# === MAIN DRIVER ===
def process_in_batches(df):
    """Run in batches and auto-resume from last saved batch."""
    total_rows = len(df)
    num_batches = (total_rows + BATCH_SIZE - 1) // BATCH_SIZE

    for b in range(num_batches):
        start = b * BATCH_SIZE
        end = min(start + BATCH_SIZE, total_rows)
        batch_file = os.path.join(BATCH_FOLDER, f"batch_{start}_{end}.csv")

        # skip already completed batches
        if os.path.exists(batch_file):
            print(f"✅ Skipping batch {b+1}/{num_batches} ({start}-{end}) — already done.")
            continue

        print(f"🚀 Processing batch {b+1}/{num_batches} ({start}-{end})...")
        df_batch = df.iloc[start:end].copy()
        df_batch = fill_from_nasa(df_batch)

        # save batch progress
        df_batch.to_csv(batch_file, index=False)
        print(f"💾 Saved {batch_file}")

    print("🎉 All batches complete. You can now merge them using combine_batches().")


# === COMBINE ===
def combine_batches(output_file="asteroids_full_filled.csv"):
    """Combine all saved batches into a single CSV."""
    import glob
    batch_files = sorted(glob.glob(os.path.join(BATCH_FOLDER, "batch_*.csv")))
    df_full = pd.concat([pd.read_csv(f) for f in batch_files], ignore_index=True)
    df_full.to_csv(output_file, index=False)
    print(f"✅ Combined file saved as {output_file}")


# === USAGE EXAMPLE ===
# df = pd.read_csv("your_dataset.csv")

```

```python
process_in_batches(df)
combine_batches()
```

```python
df['Relative Velocity km per hr'] = pd.to_numeric(df['Relative Velocity km per hr'], errors='coerce')

df['Orbital Period'] = pd.to_numeric(df['Orbital Period'], errors='coerce')

valid = df['Orbital Period'].dropna()
bins = pd.qcut(valid, q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

df.loc[valid.index, 'Orbital Period Category'] = bins

text_mapping = {'Low': 'Q1', 'Medium': 'Q2', 'High': 'Q4'}
df['Orbital Period Category'] = df['Orbital Period Category'].combine_first(
    df['Orbital Period'].replace(text_mapping)
)
quartile_map = {'Q1': 1, 'Q2': 2, 'Q3': 3, 'Q4': 4}
df['Orbital Period Category'] = df['Orbital Period Category'].map(quartile_map)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

features = ['Relative Velocity km per hr', 'Miss Dist.(Astronomical)',
            'Jupiter Tisserand Invariant', 'Semi Major Axis', 
            'Asc Node Longitude', 'Perihelion Arg', 'Aphelion Dist', 
            'Mean Anomaly', 'Mean Motion', 'Orbital Period Category']
target = 'Hazardous'

df = df.dropna(subset=[target])

numeric_features = df[features].select_dtypes(include=['number']).columns

df[numeric_features] = df[numeric_features].fillna(df[numeric_features].mean())

categorical_features = [col for col in features if col not in numeric_features]
for col in categorical_features:
    df[col] = df[col].fillna(df[col].mode()[0])

df[target] = df[target].astype(int)

X_train, X_test, y_train, y_test = train_test_split(df[features], df[target],
                                                    test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

model = RandomForestClassifier(n_estimators=200, random_state=42, class_weight='balanced')
model.fit(X_train_scaled, y_train)

y_pred = model.predict_proba(X_test_scaled)[:, 1]
threshold = 0.3  
y_pred_custom = (y_pred >= threshold).astype(int)
print(confusion_matrix(y_test, y_pred_custom))
print(classification_report(y_test, y_pred_custom))
```

```python
importances = model.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 6))
plt.barh(range(len(indices)), importances[indices], align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Feature Importance')
plt.title('Which Features Predict Hazardous Asteroids')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc, precision_recall_curve, f1_score

y_proba = model.predict_proba(X_test_scaled)[:, 1]

fpr, tpr, roc_thresholds = roc_curve(y_test, y_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate (Recall)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()

precision, recall, pr_thresholds = precision_recall_curve(y_test, y_proba)
f1_scores = 2 * (precision * recall) / (precision + recall + 1e-8)
best_idx = np.argmax(f1_scores)
best_threshold = pr_thresholds[best_idx]
best_f1 = f1_scores[best_idx]

plt.figure(figsize=(6, 5))
plt.plot(recall, precision, label='Precision–Recall Curve')
plt.scatter(recall[best_idx], precision[best_idx], color='red', label=f'Best F1 = {best_f1:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision–Recall Curve')
plt.legend()
plt.show()

print(f"🔹 Best threshold for maximum F1-score: {best_threshold:.2f}")
print(f"🔹 F1-score at this threshold: {best_f1:.2f}")

y_pred_custom = (y_proba >= best_threshold).astype(int)
print("\nClassification Report with Optimal Threshold:")
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred_custom))
print(classification_report(y_test, y_pred_custom))
```
